```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

wages=pd.read_stata('wages (version 13).dta')
# We will look at the first ten rows to get to know the data
wages.head(10)
```

```
# We will rename the variables female to gender and wages to wages_per_hour, and replace the 0s with Male and 1s with Female
wages = wages.rename(columns={'female':'gender','wages':'wages_per_hour_USD'})
wages['gender']=wages['gender'].map({0.0:'Male',1.0:'Female'})
wages.head()
```

```
# Let's first describe our data to check its validity
wages.describe()
```

```
ax=sns.boxplot(data=wages,orient='h',palette='Set2')
```

```
wages.wages_per_hour_USD.dtype
```

```
sns.barplot(data=wages.sort_values('wages_per_hour_USD'),y='occupation',x='wages_per_hour_USD',hue = 'gender',\
            ci=None, order = ['Management','Professional','Sales','Other','Clerical','Service'], palette = "Blues_d")
```

```
sns.catplot(data=wages.sort_values('wages_per_hour_USD'),y='occupation',x='wages_per_hour_USD',hue = 'south',\
            ci=None, order = ['Management','Professional','Sales','Other','Clerical','Service'], palette = "Blues_d",kind='bar',col='gender')
```

```
sns.catplot(data=wages.sort_values('wages_per_hour_USD'),y='occupation',x='wages_per_hour_USD',hue = 'married',\
            ci=None, order = ['Management','Professional','Sales','Other','Clerical','Service'], palette = "Blues_d",kind='bar',col='gender')
```

```
sns.displot(wages, x='wages_per_hour_USD',binwidth=3,hue='occupation',col='gender',element='step')
```

```
sns.displot(wages, x='wages_per_hour_USD',binwidth=3,hue='gender',col='occupation',element='step',stat='probability',common_norm=False)
```

```
sns.displot(wages,x='wages_per_hour_USD',kde='line',col='gender',hue='unionmember')
```

```
sns.displot(wages,x='wages_per_hour_USD',y='age',col='occupation',hue='gender',kind='kde',fill=True,rug=False)
```
