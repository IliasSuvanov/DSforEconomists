```python
tsla_df = yf.download('TSLA', 
                      start='2019-01-01', 
                      end='2019-12-31', progress=False)
tsla_df.head(10)
```

```python
high_prices = list(tsla_df['High'])
```

```python
high_prices[0] = 80
```

```python
low_prices = list(tsla_df['Low'])
```

```python
close_prices = list(tsla_df['Low'])
```

```python
plt.xlabel("days")
plt.ylabel("dollars")
plt.title("Stock prices of TSLA")

plt.plot(high_prices, label = "High")
plt.plot(low_prices, label = "Low")
plt.plot(close_prices, label = "Close")
plt.legend(loc = "best")
```

```python
appl_df = yf.download('AAPL', 
                      start='2019-01-01', 
                      end='2019-12-31', 
                      progress=False)
```

```python
plt.xlabel("days")
plt.ylabel("dollars")
plt.title("Stock prices of AAPL and TSLA")

plt.style.use('Solarize_Light2')
plt.grid(True)
plt.plot(appl_df['Adj Close'],color='red', label="Apple", linestyle = "dashed")
plt.plot(tsla_df['Adj Close'],color = 'blue', label = "Tesla")
plt.legend(loc="upper left")
```

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "https://www.newtraderu.com/wp-content/uploads/2020/06/1200px-Candlestick_chart_scheme_03-en.svg-1024x683.png")
```

```python
data_appl = appl_df[['Open', 'High', 'Low','Close']]
```

```python
data_appl.head()
```

```python
data_appl.reset_index(inplace = True)
```

```python
data_appl.head()
```

```python
data_appl['Date'] = data_appl['Date'].map(mdates.date2num)
```

```python
data_appl.head()
```

```python
ax = plt.subplot()
ax.grid(True)
ax.set_axisbelow(True)
ax.xaxis_date()
ax.set_title("Apple stock chart", color = "white")
ax.set_facecolor('black')
ax.figure.set_facecolor('grey')
ax.tick_params(axis = 'x', colors = 'white')
ax.tick_params(axis = 'y', colors = 'white')
candlestick_ohlc(ax, data_appl.values, width = 1, colorup = '#61de2a')
plt.show()
```
